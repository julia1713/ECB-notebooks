### Public data taken from the SBTi website: https://sciencebasedtargets.org/download/excel
    
The Science Based Targets initiative (SBTi) is a partnership between CDP (a not-for-profit charity), the United Nations Global Compact, World Resources Institute (WRI) and the World Wide Fund for Nature (WWF). 


The SBTi drives ambitious climate action in the private sector by enabling organizations to set science-based emissions reduction targets.

```python
query2= """
query to get all LEIs"""
all_LEIs = pd.read_sql_query(query2, cnxn)
```

```python
all_LEIs = all_LEIs.rename(columns={'entty_riad_cd': 'RIAD code', 'entty_cd': 'LEI'})  
all_LEIs
```

```python
companies = pd.read_excel('C:/Users/user/Desktop/tasks/companies-taking-action.xlsx')
companies
```

```python
companies = companies[companies['LEI'].notna()]
companies
```

```python
all_LEIs['entty_cd'].isin(companies['LEI']).value_counts() #87% of LEIs in companies-taking-action excel are in RIAD
```

```python
all_in_RIAD = companies[companies['LEI'].isin(all_LEIs['entty_cd'])]
all_in_RIAD
```

```python
merged = pd.merge(all_in_RIAD, all_LEIs, on=['LEI'], how='inner')
merged
```

```python
merged = merged.rename(columns={'Location': 'Country'})  

merged = merged[['LEI', 'RIAD code', 'Company Name', 'Country', 'Sector']]
merged.to_excel('in_RIAD.xlsx', index = False)
```
